```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your uploaded dataset
df = pd.read_csv("/content/train.csv")

df.head()
```

```python
plt.figure()
sns.barplot(x="Sex", y="Survived", data=df)

plt.title("Survival Rate by Gender")
plt.ylabel("Survival Rate")
plt.xlabel("Gender")
plt.ylim(0,1)  # Honest baseline
plt.show()
```

```python
plt.figure()
sns.barplot(x="Pclass", y="Survived", data=df)

plt.title("Survival Rate by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
plt.ylim(0,1)
plt.show()
```

```python
plt.figure()
sns.histplot(df["Age"].dropna(), bins=20)

plt.title("Age Distribution of Titanic Passengers")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure()
sns.histplot(data=df, x="Age", hue="Survived", bins=20, element="step")

plt.title("Age Distribution by Survival Status")
plt.xlabel("Age")
plt.ylabel("Count")
plt.legend(title="Survived", labels=["No", "Yes"])
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/train.csv")

plt.figure()
sns.barplot(x="Embarked", y="Survived", data=df)

plt.title("Survival Rate by Port of Embarkation")
plt.xlabel("Embarkation Port (C = Cherbourg, Q = Queenstown, S = Southampton)")
plt.ylabel("Survival Rate")
plt.ylim(0,1)
plt.show()
```

```python
plt.figure()
sns.boxplot(x="Survived", y="Fare", data=df)

plt.title("Fare Distribution by Survival")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Fare Paid")
plt.show()
```

```python
plt.figure()
sns.scatterplot(x="Age", y="Fare", hue="Survived", data=df)

plt.title("Age vs Fare by Survival")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

# MISLEADING VISUALIZATION

```python
plt.figure()
sns.barplot(x="Pclass", y="Survived", data=df)

plt.title("Survival Rate by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
plt.ylim(0.3,0.7)  # Misleading scale
plt.show()
```

```python
adults = df[df["Age"] > 18]

plt.figure()
sns.barplot(x="Sex", y="Survived", data=adults)

plt.title("Survival Rate by Gender (Adults Only)")
plt.xlabel("Gender")
plt.ylabel("Survival Rate")
plt.ylim(0,1)
plt.show()
```

```python
# only for young passengers
young = df[df["Age"] < 25]

plt.figure()
sns.histplot(young["Age"].dropna(), bins=10)

plt.title("Age Distribution (Passengers Under 25)")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```
