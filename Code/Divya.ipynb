```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


```

```python
df = pd.read_csv("/content/train.csv")

df.head()
```

#### We are using train.csv because it contains the survival outcome variables(0=did not durvive,1=survived), which make easy to analyze patterns and create both ethical and misleading visualizations.

# ACCURATE VISUALIZATIONS

# Visualization By gender

```python
plt.figure()
sns.barplot(x="Sex", y="Survived", data=df)

plt.title("Survival Rate by Gender")
plt.ylabel("Survival Rate")
plt.xlabel("Gender")
plt.ylim(0,1)
plt.show()
```

###### This visualization compares survival rates between male and female passengers.Out of 577 male passengers, only 109 survived (18.9%). In contrast, 233 out of 314 female passengers survived (74.2%). This large difference highlights the strong impact of gender on survival outcomes.

# Visulization by Passenger class

```python
plt.figure()
sns.barplot(x="Pclass", y="Survived", data=df)

plt.title("Survival Rate by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
plt.ylim(0,1)
plt.show()
```

###### This visualization compares survival rates across passenger classes.Out of 216 first-class passengers, 136 survived (63.0%).Among 184 second-class passengers, 87 survived (47.3%).In contrast, only 119 out of 491 third-class passengers survived (24.2%).

# Visualization of Survival rate of Embarkation Port

```python
plt.figure()
sns.barplot(x="Embarked", y="Survived", data=df)

plt.title("Survival Rate by Port of Embarkation")
plt.xlabel("Embarkation Port (C = Cherbourg, Q = Queenstown, S = Southampton)")
plt.ylabel("Survival Rate")
plt.ylim(0,1)
plt.show()
```

##### This visualization compares survival rates based on the port where passengers boarded the Titanic.Passengers who boarded at Cherbourg had the highest survival rate(55.4%), followed by Queenstown, while Southampton passengers had the lowest survival rate (39.0%),(33.7%), i think the embarkation location influenced survival outcomes.

# Visualization of survivals by the fare Distribution

```python
plt.figure()
sns.boxplot(x="Survived", y="Fare", data=df)

plt.title("Fare Distribution by Survival")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Fare Paid")
plt.show()
```

###### This boxplot compares the fare paid by passengers who did not survive (0) and those who survived (1).By the plot we can say that Passengers who survived generally paid higher fares than those who did not survive.
